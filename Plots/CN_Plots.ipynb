```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel("CN_loss.xlsx")
```

```python
data
X = data['Loss Percentage']
Y = data['Throughput']
X=X*100
```

```python
plt.plot(X,Y,marker='o')
plt.xlabel("Packet Loss in %")
plt.ylabel("Throughput in bytes/second")
plt.title("Packet Loss vs Throughput")
```
